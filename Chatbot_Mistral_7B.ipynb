```python
from transformers import pipeline

model_name = "deepset/roberta-base-squad2"

# Initialize the question-answering pipeline
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)

# Ask the user to enter a question
user_question = input("Enter your question: ")

# Define the context
context = 'The option to convert models between FARM and transformers gives freedom to the user and let people easily switch between frameworks.'


while True:
    # Ask the user to enter a question or type 'exit' to quit
    user_question = input("Enter your question (or type 'exit' to quit): ")

    if user_question.lower() == 'exit':
        break  # Exit the loop if the user types 'exit'

    # Create the QA input dictionary with the user's question and the predefined context
    QA_input = {
        'question': user_question,
        'context': context
    }

    # Get the bot's answer based on the user's question and context
    res = nlp(QA_input)

    # Print the bot's answer
    print("Bot's Answer:", res['answer'])
```

```python
from langchain.chains import RetrievalQA
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import LlamaCpp
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.document_loaders import PyPDFDirectoryLoader
```

```python
#load pdf files
loader = PyPDFDirectoryLoader("/content/sample_data/Data/")
data = loader.load()
```

```python
print(data)
```

```python
#Step 05: Split the Extracted Data into Text Chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=20)

text_chunks = text_splitter.split_documents(data)
```

```python
len(text_chunks)
```

```python
#get the third chunk
text_chunks[2]
```

```python
#Step 06:Downlaod the Embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
#Step 08: Create Embeddings for each of the Text Chunk
vector_store = FAISS.from_documents(text_chunks, embedding=embeddings)
```

```python
#connect to google drive
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
#Import Model
llm = LlamaCpp(
    streaming = True,
    model_path="/content/drive/MyDrive/mistral-7b-instruct-v0.1.Q3_K_S.gguf",
    temperature=0.75,
    top_p=1,
    verbose=True,
    n_ctx=4096
)
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector_store.as_retriever(search_kwargs={"k": 2}))
```

```python
query = "Give information about dell venue tablets?"
```

```python
qa.run(query)
```

```python
import sys

while True:
  user_input = input(f"Input Prompt: ")
  if user_input == 'exit':
    print('Exiting')
    sys.exit()
  if user_input == '':
    continue
  result = qa({'query': user_input})
  print(f"Answer: {result['result']}")
```
